```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29
df=pd.read_csv('data_cancer.csv')

df.head()
```

```python
df.isna().sum()
```

```python
df=df.drop(['Unnamed: 32','id'],axis=1)
df.shape

#convert B and M into 0 and 1, otherwise random forest cannot process
df['diagnosis'].mask(df['diagnosis'] == 'B', 0, inplace=True)
df['diagnosis'].mask(df['diagnosis'] == 'M', 1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

X=df.iloc[:,1:33]
Y=df.iloc[:,0]
X_train,X_test,y_train,y_test=train_test_split(X,Y,random_state=1)



#make sure Y will be the output results
Y.head()
```

```python
randomForest = RandomForestRegressor(n_estimators = 1000, random_state = 42)
randomForest.fit(X_train, y_train)

```

```python
randomForest.score(X_test,y_test)
```

```python
#random select 4th row from X_test, note: X_test has no result
df_predict=X_test.iloc[10:11,:]
df_predict.shape
df_predict.head()
```

```python
#produce output with input of df_predict
prediction_test=randomForest.predict(df_predict)

if prediction_test>0.9:
  print("Result is: Maglinant")
elif prediction_test<0.1:
  print("Result is: Benign")
else:
  print("Error: not within range")
```
